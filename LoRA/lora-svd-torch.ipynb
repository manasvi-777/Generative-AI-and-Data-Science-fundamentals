```python
import torch
import numpy as np
_=torch.manual_seed(0)  #basically outputs the same output every time
```

```python
#Initializing dimensions of A and B matrices
d=10
k=10
r=2 #r<<mmin(d,k)

```

```python
#W->actual weights matrix(d,k)
W=torch.randn(d,r)@torch.randn(r,k)
print(W)
```

```python
W_rank=np.linalg.matrix_rank(W)   #number of linearly independent rows or columns in a matrix.
print(f'Rank of W is {W_rank}')
```

Single Variable Decomposition (SVD) of W

```python
U,S,V=torch.svd(W)

U_r=U[:,:W_rank]
S_r=torch.diag(S[:W_rank])
V_r=V[:,:W_rank].t()

B=U_r@S_r
A=V_r

print(f'Shape of A: {A.shape}')
print(f'Shape of B: {B.shape}')
```

```python
bias=torch.randn(d)
x=torch.rand(d)

y=W@x+bias
y2=(B@A)@x+bias

print("Original y using W:\n",y)
print("Total parameters of W:",W.nelement())
print()
print("y2 using BA",y2)
print("Total parameters of B and A:",B.nelement()+A.nelement())
```

Thus we see that the values are exactly the same
